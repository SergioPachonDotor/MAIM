```python
medios_de_cultivo = set(df['medio'])
datos_refinados = [df['crecimiento'][df['medio'] == medio] for medio in medios_de_cultivo]
```

```python
datos_medio_C  = datos_refinados[0]
datos_medio_OF = datos_refinados[1]
datos_medio_DH = datos_refinados[2]
datos_medio_IF = datos_refinados[3]
```

Kolmogorov Smirnov asume que $H_{0}$ sigue una distribucin normal y $H_{1}$ es diferente de una distribucin normal.
$ \alpha = 0.05 $

```python
indices = [f'{v}' for k,v in enumerate(['Medio C', 'Medio DH', 'Medio IF', 'Medio OF'])];
medios_arr_df = [datos_medio_C, datos_medio_DH, datos_medio_IF, datos_medio_OF]
for i in range(len(medios_arr_df)):
    print('\n')
    print(indices[i])
    print('----'*20)
    print(f'- Longitud de datos: {len(medios_arr_df[i])}')
    print(f'- KS: {stats.kstest(medios_arr_df[i], "norm")}')
    print(f'- Shapiro: {stats.shapiro(medios_arr_df[i])}')
```
